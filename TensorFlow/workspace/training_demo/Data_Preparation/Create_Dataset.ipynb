# Imports

```python
import os
import cv2
import numpy as np
import pandas as pd
import glob
import random
from PIL import Image, ImageFilter
```

# Create Blank Image of size 1024 * 1024

```python
def Create_Blank():   
    # white blank image
    blank_image = 255 * np.ones(shape=[1024, 1024, 3], dtype=np.uint8)    
    cv2.imwrite('BlankImage.jpg',blank_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return blank_image 
```

```python
def Change_Orientation(img, angle):   
    rot_img = img.rotate(angle)    
    rot_img = rot_img.filter(ImageFilter.GaussianBlur(3))
#     rot_img.show()
    return rot_img    
```

# Randomly place symbols in Blank Image

```python
dataset_dir = os.path.join(os.getcwd(), "Dataset\\")
images = glob.glob(os.getcwd() + r'\Symbols\*.jpg')
useHeader = True
i = 1
while i < 121:
    img_bg = Create_Blank()
    img_bg = Image.open("BlankImage.jpg")
    width, height = img_bg.size
    base_name = os.path.basename(img_bg.filename)
    for img in images:
        img = Image.open(img)
        small_img_name = os.path.basename(img.filename)
#         print(type(small_img_name))
        small_img_name = small_img_name.rsplit('.', 1)[0]
#         print(small_img_name)
        w, h = img.size
#         print(w,h)
        x = random.randint(40, width-40)
        y = random.randint(40, height-40)
        x = x+10
        y = y+10
        if ((x+w) <= 1024 and (y+h) <= 1024):
            img_bg.paste(img, (x, y))
            df = pd.DataFrame(data={"filename": [str(i) + ".jpg"], "width": [width], "height": [height], "class": [small_img_name], "xmin": [x], "ymin": [y], "xmax": [x+w], "ymax": [y+h]})
            df.to_csv("./labels.csv", sep=',', doublequote=False, index=False, mode="a", header = useHeader)
            useHeader = False  
        else:
            print("Out of boundary, so skipping the Image" + small_img_name)
    img_bg.save(dataset_dir + str(i) + '.png', 'PNG')   
    print("Saved " + str(i) + " image")
    img_bg.close()
    if os.path.exists(os.getcwd() + "\\" + base_name):
#         print("Closed and removing blank image " + str(i))
        os.remove(os.getcwd() + "\\" + base_name)        
    i += 1
print("Completed!!!")
```
